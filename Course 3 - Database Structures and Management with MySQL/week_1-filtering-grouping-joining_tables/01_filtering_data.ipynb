## Filtering data using AND, OR, and NOT logical operators

In this notebook, we will learn how to filter data using the AND, OR, and NOT logical operators. We will use the `SELECT` statement to retrieve data from the `Lucky_Shrub` database. We will also use the `WHERE` clause to filter data based on a specified condition.

Let's Explore the syntax of each of these logical operators.

#### AND logical operator

The `AND` logical operator is used to combine two or more conditions in a `WHERE` clause. It retrieves rows where all the conditions are true. The syntax for the `AND` logical operator is as follows:

```sql
SELECT column1, column2, ...
FROM table_name
WHERE condition1 AND condition2 AND condition3 ...;
```

#### OR logical operator

The `OR` logical operator is used to combine two or more conditions in a `WHERE` clause. It retrieves rows where at least one of the conditions is true. The syntax for the `OR` logical operator is as follows:

```sql
SELECT column1, column2, ...
FROM table_name
WHERE condition1 OR condition2 OR condition3 ...;
```

#### NOT logical operator

The `NOT` logical operator is used to negate a condition in a `WHERE` clause. It retrieves rows for which the condition is false. The syntax for the `NOT` logical operator is as follows:

```sql
SELECT column1, column2, ...
FROM table_name
WHERE NOT condition;
```

## Filtering data using IN, NOT IN, and BETWEEN, Like logical operators

In this notebook, we will learn how to filter data using the `IN`, `NOT IN`, and `BETWEEN` logical operators. We will use the `SELECT` statement to retrieve data from the `Lucky_Shrub` database. We will also use the `WHERE` clause to filter data based on a specified condition.



Let's Explore the syntax of each of these logical operators.

#### IN logical operator

The `IN` logical operator is used to specify a range of values. It retrieves rows where the specified column is equal to any value in a list. The syntax for the `IN` logical operator is as follows:

```sql
SELECT column1, column2, ...
FROM table_name
WHERE column_name IN (value1, value2, ...);
```

**Example - IN** 

Using a Country table, the following SQL statement selects all customers from the country 'Germany', 'France', and 'UK':

```sql
SELECT *
FROM Customers
WHERE Country IN ('Germany', 'France', 'UK');
```

#### NOT IN logical operator

The `NOT IN` logical operator is used to specify a range of values. It retrieves rows where the specified column is not equal to any value in a list. The syntax for the `NOT IN` logical operator is as follows:

```sql
SELECT column1, column2, ...
FROM table_name
WHERE column_name NOT IN (value1, value2, ...);
```

**Example - NOT IN**

Using a Country table, the following SQL statement selects all customers except those from the country 'Germany', 'France', and 'UK':

```sql
SELECT *
FROM Customers
WHERE Country NOT IN ('Germany', 'France', 'UK');
```

#### BETWEEN logical operator

The `BETWEEN` logical operator is used to retrieve rows based on a range of values. The syntax for the `BETWEEN` logical operator is as follows:

```sql
SELECT column1, column2, ...
FROM table_name
WHERE column_name BETWEEN value1 AND value2;
```

**Example - BETWEEN**

Using a Customers table, the following SQL statement selects all customers with a `CustomerID` between 1 and 10:

```sql
SELECT *
FROM Customers
WHERE CustomerID BETWEEN 1 AND 10;
```

#### LIKE logical operator

The `LIKE` logical operator is used to retrieve rows based on a pattern. The syntax for the `LIKE` logical operator is as follows:

```sql
SELECT column1, column2, ...
FROM table_name
WHERE column_name LIKE pattern;
```

**Example - LIKE**

Using a Customers table, the following SQL statement selects all customers with a `CustomerName` that starts with 'a':

```sql
SELECT *
FROM Customers
WHERE CustomerName LIKE 'a%';
```

Reference sheet: Using operators and the WHERE clause
Use the WHERE clause in SQL SELECT, UPDATE and DELETE statements to filter data.

The WHERE clause can specify one or more filter conditions to determine how data is limited or narrowed down from the queried tables.

It's also possible to use logical operators and some other operators when formulating filter conditions.

The logical AND and OR operators are used as conjunctive operators to combine multiple conditions in the WHERE clause. The logical NOT operator negates the result of evaluating a given condition (or conditions) and filters data accordingly. 

The IN operator can be an alternative to the OR operator. The IN operator can check multiple values, equivalent to using the OR logical operator to specify various conditions. 

The BETWEEN operator is a convenient way to perform a range check on numeric and date values. It can be used as an alternative to greater than or equal to (>=) and less than or equal to (<=) comparison operators. 

Use the LIKE operator to match and filter data based on these patterns. The wildcard characters, percentage (%) and underscore (_) are used to formulate patterns. The percentage ( % ) wildcard matches any string of zero or more characters. The underscore (_) wildcard matches any single character.
